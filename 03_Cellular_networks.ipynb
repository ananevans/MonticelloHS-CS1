# Intro to Cellular Networks


## Components of a Cellular Network

### Components

1. Mobile Switching Center 
  1. Connects calls between subscribers 
  1. Maintains a database of subscribers and the mobile phones current locations.
2. Cell Towers
  1. Communicate with the mobile phones in their range.
  1. Connect to the MSC though fiber optic cables
3. Mobile Phone  

### Alice's phone finds Bob's phone

1. Alice's phone makes a request to the cell tower.
2. The cell tower passes the call request to the MSC.
3. The MSC locates the cell tower serving Bob's phone.
4. The MSC passes the call request to the cell tower.
5. The cell tower forwards the call request to Bob's phone.

## Alice talks to Bob (Roaming)

1. Alice says *Hello!*
2. Alice's mobile phone transforms voice to data and sends using radio waves to the **Cell Tower** with the strongest signal.
3. The **Cell Tower** converts the data and sends it over a fiber optic cable to a **Message Switching Center** it is connected to.
4. The **Message Switching Center** transmits the data to another **Message Switching Center** over internet or a private netwok.
5. That second **Message Switching Center** transmits the data over a fiber optic cable to a **Cell Tower**.
6. The **Cell Tower** transmits data using radio waves to Bob's phone.
7. Bob's phone converts the data to an audible signal.
8. Bob hears *Hello!*

## Alice's phone finds Bob's phone (Roaming)


1. Alice's phone is connected to an MSC different than the one it is registered with. 
1. Alice's phone provides information on its **Home MSC**
2. Alice's phone is recorded as a visiting device with the **Visitor MSC**.
3. The **Visitor MSC** informs Alice's **Home MSC** that it services Alice's calls.
4. The same happens with Bob's phone.
1. Alice calls Bob.
2. Alice's **Visitor MSC** forwards the call to Bob's **Home MSC**.
1. Bob's **Home MSC** tells Alice's **Visitor MSC** that Bob's phone is serviced by Bob's **Visitor MSC**.
1. Alice's **Visitor MSC** routes the call request to Bob's **Visitor MSC**.

## Activity

The students are organized in groups of 3 or 4 students.

1. Each group member chooses a role:
  1. MSC (one student)
  2. Cell tower (one student)
  3. Mobile phone (one or two students)
2. The MSC records the phones in the group. This will be the home MSC for the phone.
2. One phone from each group roams to the tower of another group.
  1. The phone announces its presence to the tower.
  2. The tower send the information to the MSC.
  3. The MSC records the new phone as a visiting.
  4. The MSC sends a message to the phone's home MSC
  5. The phone' home MSC records the vising MSC.
3. One visiting phone chooses a phone at random to call. 
4. Record the steps needed to establish connection.
