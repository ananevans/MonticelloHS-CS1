### Binary Search

In class, we worked out that a guesser needs at most 7 tries to find the answer when they use the strategy to always guess the midpoint. Here is why:
1. At the start of the game the interval in which the secret number falls has length 100.
1. After the first guess, the interval in which the secret number falls has length 50.
1. After the second guess, the interval in which the secret number falls has length 25.
1. After the third guess, the interval in which the secret number falls has length 12.
1. After the fourth guess, the interval in which the secret number falls has length 6.
1. After the fifth guess, the interval in which the secret number falls has length 3.
1. After the sixth guess, the interval in which the secret number falls has length 1.
1. For the seventh guess, there is only one choice left and the game is over.
